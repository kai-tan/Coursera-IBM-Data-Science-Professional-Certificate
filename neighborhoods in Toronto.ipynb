```python
from pandas.io.html import read_html 
page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

wikitables = read_html(page, attrs={"class":"wikitable"})

wikitables[0].head()
```

```python
wikidataframe = wikitables[0]
wikidataframe = wikidataframe[wikidataframe['Borough'] != 'Not assigned']
```

```python
# wikidataframe = wikidataframe.reset_index()
wikidataframe = wikidataframe[["Postcode", "Borough", "Neighbourhood"]]
wikidataframe.shape
```

```python
df2= wikidataframe.groupby('Postcode').agg(lambda x: ', '.join(x))
df2[df2["Neighbourhood"] == 'Not assigned']
```

```python
df2.loc[df2['Neighbourhood']=="Not assigned",'Neighbourhood'] = df2.loc[df2['Neighbourhood']=="Not assigned",'Borough']
```

```python
df2
```

```python
df3 = df2.reset_index()
df3
```

Remove duplication in Borough column 

```python
df3['Borough']= df3['Borough'].str.replace('nan|[{}\s]','').str.split(',').apply(set).str.join(',').str.strip(',').str.replace(",{2,}",",")
```
